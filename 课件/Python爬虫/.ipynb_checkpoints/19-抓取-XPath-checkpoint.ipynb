### Python使用XPath

首先我们使用 lxml 的 etree 库，然后利用 etree.HTML 初始化，然后我们将其打印出来。

其中，这里体现了 lxml 的一个非常实用的功能就是自动修正 html 代码，大家应该注意到了，最后一个 li 标签，其实我把尾标签删掉了，是不闭合的。不过，lxml 因为继承了 libxml2 的特性，具有自动修正 HTML 代码的功能。

#### 文件读取内容：

利用 parse 方法来读取文件。

```python
from lxml import etree
html = etree.parse('hello.html')
result = etree.tostring(html, pretty_print=True)
print(result.decode('utf-8'))
```

#### XPath实例

获取所有的 <li> 标签

```python
from lxml import etree
html = etree.parse('hello.html')
print(type(html))
result = html.xpath('//li')
print(result)
print(len(result))
print(type(result))
print(type(result[0]))
```

```python
from lxml import etree
html = etree.parse('hello.html')
result = html.xpath('//li/@class')
print(result)
```

```python
from lxml import etree
html = etree.parse('hello.html')
result = html.xpath('//li/a[@href="link1.html"]')
print(result)
```

```python
# -*- coding:utf-8 -*-
from lxml import etree
html = etree.parse('hello.html')

#注意这么写是不对的,因为 / 是用来获取子元素的，而 <span> 并不是 <li> 的子元素，所以，要用双斜杠
#result = html.xpath('//li/span')
result = html.xpath('//li//span')
print(result)
```

```python
# -*- coding:utf-8 -*-
from lxml import etree
html = etree.parse('hello.html')
result = html.xpath('//li//span/text()')
print(result)
```
